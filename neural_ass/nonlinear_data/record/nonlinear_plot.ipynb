```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
revenue_list1 = np.load('mle_true_revenue_list1.npy').tolist() 
revenue_list2 = np.load('mle_simulate_revenue_list2.npy').tolist()
true_profit_list = np.load('mle_purchase_revenue_list.npy').tolist() 
```

```python
differ_list = [revenue_list1[i] - revenue_list2[i] for i in range(len(revenue_list1))]
differ_list2 = [revenue_list1[i] - true_profit_list[i] for i in range(len(revenue_list1))]
```

```python
cumulative_sum1 = []
current_sum = 0
for num in differ_list:
    current_sum += num
    cumulative_sum1.append(current_sum)

cumulative_sum2 = []
current_sum = 0
for num in differ_list2:
    current_sum += num
    cumulative_sum2.append(current_sum)
```

```python
# alpha = 0.01  lambda = 1 model(5,10,10,2) [10,20,20,10] adam(lr = 0.01, decay = 0.01)
 
```

```python
plt.figure(figsize=(10,12),dpi = 100)
plt.subplot(3,1,1)
plt.title('alpha = 0.01  lambda = 1 model(5,10,10,4) [10,20,20,20,20,10] adam(lr = 0.001, decay = 0)')
plt.plot(revenue_list1,label = 'true')
plt.plot(revenue_list2, label = 'simulate')
plt.plot(true_profit_list,label = 'purchase')
plt.xlim(0,10000)
plt.legend()

plt.subplot(3,1,2)
plt.title("expected true - purchase cumulative sum")
plt.plot(cumulative_sum2,label = 'cum_diff')
plt.xlim(0,10000)

plt.subplot(3,1,3)
plt.title("expected true - simulation cumulative sum")
plt.plot(cumulative_sum1,label = 'cum_diff')
plt.xlim(0,10000)

plt.show()
```
