```python
import numpy as np
import matplotlib.pyplot as plt

# 生成1000个随机起点和终点
start_points = np.linspace(0, 10, 101)
end_points = start_points[-100:]
end_points
```

```python
lista = []
for i in range(100):
    lista.append(np.random.uniform(start_points[i],end_points[i],10))
lista
```

```python
np.array(random.sample(lista,8))
```

```python
features = [np.array(random.sample(lista,8)) for i in range(10000)]
```

```python
theta = np.random.rand(10, 1).reshape(1,-1)
# theta /10 
theta = theta / 5
theta
```

```python
features[0]
```

```python
features[0][0] @theta.T
```

```python
features[0] @ theta.T
```

```python
ut_list = [(features[i] @ theta.T).reshape(1,-1) for i in range(10000)]
```

```python
ut_list
```

```python
np.save('theta_gau.npy', theta)
np.save('features_uni.npy', features)
```

```python
def prob(vj_list):
    sum = np.sum(np.exp(vj_list)) 
    return [np.exp(vj_list[i]) / sum for i in range(len(vj_list))]  
```

```python
prob_list = [prob(ut_list[i])[0] for i in range(10000)]
prob_list 
```

```python
import random
def get_purchase(problist):
    returnlist = [0 for i in range(len(problist))]
    indexchoose = random.choices([i for i in range(len(problist))], weights = problist)[0]
    returnlist[indexchoose] = 1
    return np.array(returnlist)
```

```python
purchase_list = []
for i in range(10000):
    purchase_list.append(get_purchase(prob_list[i]))
```
